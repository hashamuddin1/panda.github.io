## **FILTERING**

```python
#TO PRINT ANY PARTICULAR COLUMN
a.loc[:,"NAME"]
```

```python
a.loc[a["GRADES"]=="A"]
```

```python
a.loc[a["MARKS"]>75]
```

```python
#AND CONDITION
a.loc[(a["AGE"]>21) & (a["MARKS"]>75)]
```

```python
#OR CONDITION
a.loc[(a["MARKS"]<50) | (a["GRADES"]=="A")]
```

```python
a.loc[a["GRADES"]!="A"]
```
